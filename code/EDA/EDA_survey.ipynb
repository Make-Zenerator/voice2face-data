```python
def draw_percentage_histogram(title, data):
    data = list(data.dropna())
    rating = [1, 2, 3, 4, 5]
    total = len(data)
    percentage = [(data.count(x) / total) for x in rating]

    plt.bar(rating, percentage, align='center')
    plt.xlabel("1~5")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.xticks(rating)
    plt.yticks(np.arange(0, 0.4, 0.05))
    plt.grid(True)
    plt.show()
    
```

```python
title = 'Condition Image Rating'
data = result_csv['condition_image_rating']
draw_percentage_histogram(title, data)
```

```python
title = 'Voice Image Rating'
data = result_csv['voice_image_rating']
draw_percentage_histogram(title, data)
```

## 분석하기 쉽도록 merged_df 생성

```python
def extract_filename(string):
    try: 
        return string.replace("https://storage.makezenerator.com:9000/voice2face-public/", "")
    except:
        return string
    
result_csv['voice_image_url'].apply(extract_filename)
```

```python
request_csv.rename(columns = {'id':'mz_request_id'}, inplace = True)
result_csv.rename(columns = {'id':'mz_result_id'}, inplace = True)
request_csv['voice_url'] = request_csv['voice_url'].apply(extract_filename)
result_csv['condition_image_url'] = result_csv['condition_image_url'].apply(extract_filename)
result_csv['voice_image_url'] = result_csv['voice_image_url'].apply(extract_filename)
```

```python
merged_df = pd.merge(request_csv, result_csv, on='mz_request_id', how='right')
merged_df = merged_df[['mz_result_id', 'mz_request_id', 'user_id', 'age', 'gender', 'voice_url', 'condition_image_url', 'voice_image_url', 'condition_image_rating', 'voice_image_rating', 'status', 'survey']]
merged_df.head()
```

```python
merged_df.to_csv("db/merged_df.csv")
```

## 설문조사까지 포함된 full_merged_df 생성

```python
full_merged_df = pd.merge(merged_df, survey_csv, on='mz_result_id', how='right')
full_merged_df = full_merged_df.drop(['id', 'user_phone', 'created_at', 'updated_at', 'deleted_at'], axis=1)
full_merged_df.head()
```

```python
full_merged_df.to_csv("db/full_merged_df.csv")
```
